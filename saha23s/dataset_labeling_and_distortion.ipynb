|--------|-----------|-----------|------------|
|   0    |     0     |     0     |     0      |
|   1    |     1     |     0     |     0      |

```python
# Edit this code to do gaussian noise instead

def distort_data(data, distortion_factor):
    # generates Gaussian noise array with mean 0 and standard deviation as distortion_factor
    distortion = np.random.normal(loc=0.0, scale=distortion_factor, size=data.shape)
    distorted_data = np.clip(data + distortion, 0, 1)  # Clip values to the range [0, 1]
    return distorted_data


# ideas of distortion: Noise add, Outliers insert, Swap Values, Random Shuffles, Value scaling

def create_bad_dataset(good_dataset, column_names, distortion_factor):
    bad_dataset = good_dataset.copy()
    for column_name in column_names:
        column_data = bad_dataset[column_name].values
        distorted_data = distort_data(column_data, distortion_factor)
        bad_dataset[column_name] = distorted_data
    return bad_dataset
```

```python
from google.colab import files
def csv_download(final, final_name):
  file_path = '/content/drive/MyDrive/MelodyResearch/final_csv/final_normalized/' + final_name
  final.to_csv(file_path, index=False)
```

```python
# Specify the level of distortion for light, medium, and heavy noise
distortion_factors = [0.25, 0.5, 0.75]

# Specify the corresponding dataset names
dataset_names = ["push_normalized", "pull_normalized", "folding_normalized", "sponge_push_normalized"]


# List of good datasets (min-max)
arrays_of_finals_MM = [final_min_max_push_normalized_data, final_min_max_pull_normalized_data,
                    final_min_max_folding_normalized_data, final_min_max_sponge_push_normalized_data]


for i, good_dataset in enumerate(arrays_of_finals_MM):
    light = create_bad_dataset(good_dataset, columns_to_distort, 0.25)
    light["is_bad"] = 1
    light["bad_vision"] = 1
    light["bad_force"] = 1
    light["bad_robot"] = 1
    csv_download(light, f"{dataset_names[i]}_light.csv")

    medium = create_bad_dataset(good_dataset, columns_to_distort, 0.5)
    medium["is_bad"] = 1
    medium["bad_vision"] = 1
    medium["bad_force"] = 1
    medium["bad_robot"] = 1
    csv_download(medium, f"{dataset_names[i]}_medium.csv")

    heavy = create_bad_dataset(good_dataset, columns_to_distort, 0.75)
    heavy["is_bad"] = 1
    heavy["bad_vision"] = 1
    heavy["bad_force"] = 1
    heavy["bad_robot"] = 1
    csv_download(heavy, f"{dataset_names[i]}_heavy.csv")
```
