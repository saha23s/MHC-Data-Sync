```python
# Calculate the mean and standard deviation of each feature.
# standardization
def calculate_std(final):
  mean = final[columns_to_normalize].mean(axis=0)
  std_dev = final[columns_to_normalize].std(axis=0)
  normalized_data = (final[columns_to_normalize] - mean) / std_dev
  # assuming when max and min values are same - filling 0 in ?
  normalized_data = normalized_data.fillna(0)
  normalized_data['%time'] = final['%time']
  normalized_data['field.header.seq'] = final['field.header.seq']
  normalized_data['image_frames'] = final['image_frames']
  return normalized_data

final_std_push_normalized_data = calculate_std(final_push)
final_std_pull_normalized_data = calculate_std(final_pull)
final_std_folding_normalized_data = calculate_std(final_folding)
final_std_sponge_push_normalized_data = calculate_std(final_sponge_push)
```

```python
# normalization
def calculate_min_max(final):
  # Calculate the minimum and maximum values of each feature.
  min_values = final[columns_to_normalize].min(axis=0)
  max_values = final[columns_to_normalize].max(axis=0)
  # Normalize the column.
  normalized_data = (final[columns_to_normalize] - min_values) / (max_values - min_values)

  # assuming when max and min values are same - filling 0 in ?
  normalized_data = normalized_data.fillna(0)
  normalized_data['%time'] = final['%time']
  normalized_data['field.header.seq'] = final['field.header.seq']
  normalized_data['image_frames'] = final['image_frames']
  return normalized_data

final_min_max_push_normalized_data = calculate_min_max(final_push)
final_min_max_pull_normalized_data = calculate_min_max(final_pull)
final_min_max_folding_normalized_data = calculate_min_max(final_folding)
final_min_max_sponge_push_normalized_data = calculate_min_max(final_sponge_push)
```
